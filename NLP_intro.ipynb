```python
#installing all NLTK packages
#running this will result in a pop-up window secect all and download all the packages
nltk.download()
```

```python
#grabbing the webpage html
import urllib.request
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/National_Football_League')
html = response.read()
print(html)
```

```python
#to install BeautifulSoup run "!pip install beautifulsoup4" in the jupyter nootebook 
#BeautifulSoup lib is for pulling data out of HTML and XML files 
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html5lib')
text = soup.get_text(strip = True)
print(text)
```

```python
#converting text into tokens
tokens = [t for t in text.split()]
print (tokens)
```

```python
#removing stop words like [a, at, the, for...etc] we dont want them to hamper our frequency count
from nltk.corpus import stopwords
sr = stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('english'):
        
        clean_tokens.remove(token)

#counting the frequency of the words
#this is the first step towards the NLP analysis of the webpage
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items():
    print(str(key) + ':' + str(val))

```

```python
#word frequency plot

freq.plot(20, cumulative=False)
```
